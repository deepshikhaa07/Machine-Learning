*CLASSIFY AND SPLIT DATA*

```python
# classifying data
data['Species'].unique()
```

```python
data['Species']=data['Species'].replace({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
```

```python
#show data
data
```

```python
# splitting dataset
x_train, x_test, y_train, y_test = train_test_split(data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']],data['Species'], train_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

*CLASSIFIER - SVM*

```python
# SVM on Iris

clf_svm = SVC(kernel = 'linear')
clf_svm.fit(x_train,y_train)
```

```python
y_pred_svm = clf_svm.predict(x_test)
y_pred_svm
```

```python
svm_score=metrics.accuracy_score(y_test, y_pred_svm)
svm_score
```

```python
# Plot Decision Boundary

DecisionBoundaryDisplay.from_estimator(
        clf_svm,
        x_train,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=data.feature_names[0],
        ylabel=data.feature_names[1],
    )

# Scatter plot

plt.scatter(x_train[:, 0], x_train[:, 1],
            c=y_train,
            s=20, edgecolors="k")
plt.show()
```
